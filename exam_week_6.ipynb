### Question 8

```python
land_price_relationship = melbourne_housin_raw[['Landsize', 'Price']].dropna()

land_price_relationship['Landsize_cat'] = pd.qcut(land_price_relationship['Landsize'], q=7, labels=[f'Quantile {i+1}' for i in range(7)])

med_price_land_cat = land_price_relationship.groupby('Landsize_cat')['Price'].median().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Landsize_cat', y='Price', data=med_price_land_cat, palette='viridis')
plt.title('Median Property Price by Land Size Category (q=7)')
plt.xlabel('Land Size Category')
plt.ylabel('Median Price (AUD)')
plt.show()
```

### Question 9

```python
price = melbourne_housin_raw[['Price']].dropna()

plt.figure(figsize=(8, 5))
sns.boxplot(x=price['Price'], color='skyblue')
plt.title('Boxplot of Property Prices')
plt.xlabel('Price (AUD)')
plt.show()

```

```python
Q1 = price['Price'].quantile(0.25)
Q3 = price['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

non_outliers = price[(price['Price'] >= lower_bound) & (price['Price'] <= upper_bound)]

percentage_outliers = (len(price) - len(non_outliers)) / len(price) * 100

percentage_outliers
```

### Question 10

```python
import scipy.stats as stats

price_skewness = price['Price'].skew()
price_kurtosis = price['Price'].kurtosis()

plt.figure(figsize=(8, 6))
stats.probplot(price['Price'], dist="norm", plot=plt)
plt.title(f'QQ Plot of Property Prices\n(Skewness: {price_skewness:.2f}, Kurtosis: {price_kurtosis:.2f})')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid()
plt.show()

price_skewness, price_kurtosis
```

```python
region_price_relationship = melbourne_housin_raw[['Price', 'Regionname']].dropna()

median_price_by_region = region_price_relationship.groupby('Regionname')['Price'].median().reset_index()

most_expensive = median_price_by_region.loc[median_price_by_region['Price'].idxmax(), 'Regionname']
cheapest = median_price_by_region.loc[median_price_by_region['Price'].idxmin(), 'Regionname']

most_expensive, cheapest

```

### Question 14

```python
from scipy.stats import chi2_contingency

region_price_relationship['Price_Category'] = pd.cut(region_price_relationship['Price'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

contingency_table = pd.crosstab(region_price_relationship['Regionname'], region_price_relationship['Price_Category'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

n = np.sum(contingency_table.values)
cramers_v = np.sqrt(chi2 / (n * min(contingency_table.shape) - 1))

chi2, p, cramers_v
```

### Question 15

```python
melbourne_housin_raw['Date'] = pd.to_datetime(melbourne_housin_raw['Date'], format="%d/%m/%Y")

melbourne_housin_raw['Year'] = melbourne_housin_raw['Date'].dt.year

earliest_year = melbourne_housin_raw['Year'].min()
latest_year = melbourne_housin_raw['Year'].max()

median_price_earliest = melbourne_housin_raw[melbourne_housin_raw['Year'] == earliest_year]['Price'].median()
median_price_latest = melbourne_housin_raw[melbourne_housin_raw['Year'] == latest_year]['Price'].median()

percentage_change = ((median_price_latest - median_price_earliest) / median_price_earliest) * 100

earliest_year, latest_year, median_price_earliest, median_price_latest, percentage_change
```
